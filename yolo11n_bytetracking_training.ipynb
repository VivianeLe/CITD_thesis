```python
# 5) Convert UA-DETRAC XML -> YOLO format

DATA_ROOT = Path("/CITD/data")
YOLO_ROOT = Path("/CITD/ua_detrac_yolo")

IMAGES_DIR = YOLO_ROOT / "images"
LABELS_DIR = YOLO_ROOT / "labels"

for split in ["train", "val"]:
    (IMAGES_DIR / split).mkdir(parents=True, exist_ok=True)
    (LABELS_DIR / split).mkdir(parents=True, exist_ok=True)

# UA-DETRAC classes thường gặp
# Nếu annotation không có type rõ ràng, mặc định gán car
CLASS_NAMES = ["car", "bus", "van", "others"]
CLASS_MAP = {
    "car": 0,
    "bus": 1,
    "van": 2,
    "others": 3,
    "other": 3,
    "unknown": 3,
}

image_paths = list(DATA_ROOT.rglob("*.jpg")) + list(DATA_ROOT.rglob("*.png"))
xml_paths = list(DATA_ROOT.rglob("*.xml"))

print("Found images:", len(image_paths))
print("Found xml:", len(xml_paths))

# Map sequence name -> list image
seq_to_images = {}
for img_path in image_paths:
    seq_name = img_path.parent.name
    seq_to_images.setdefault(seq_name, []).append(img_path)

for k in seq_to_images:
    seq_to_images[k] = sorted(seq_to_images[k])

print("Sequences with images:", len(seq_to_images))
```

```python
# 6) Hàm parse XML UA-DETRAC

def find_image_for_frame(seq_images, frame_num):
    """
    UA-DETRAC image thường có dạng img00001.jpg.
    Nếu không đúng format, fallback theo index.
    """
    candidates = [
        f"img{frame_num:05d}.jpg",
        f"img{frame_num:05d}.png",
        f"{frame_num:05d}.jpg",
        f"{frame_num:05d}.png",
    ]
    
    name_to_path = {p.name: p for p in seq_images}
    for c in candidates:
        if c in name_to_path:
            return name_to_path[c]
    
    if 1 <= frame_num <= len(seq_images):
        return seq_images[frame_num - 1]
    
    return None


def parse_box_from_target(target):
    box = target.find("box")
    if box is None:
        return None
    
    try:
        left = float(box.attrib.get("left", 0))
        top = float(box.attrib.get("top", 0))
        width = float(box.attrib.get("width", 0))
        height = float(box.attrib.get("height", 0))
    except Exception:
        return None
    
    if width <= 0 or height <= 0:
        return None
    
    return left, top, width, height


def parse_class_from_target(target):
    attr = target.find("attribute")
    if attr is None:
        return 0
    
    vehicle_type = (
        attr.attrib.get("vehicle_type")
        or attr.attrib.get("type")
        or attr.attrib.get("vehicleType")
        or "car"
    )
    
    vehicle_type = vehicle_type.lower().strip()
    return CLASS_MAP.get(vehicle_type, 3)


def convert_xml_to_yolo(xml_path, split="train"):
    tree = ET.parse(xml_path)
    root = tree.getroot()
    
    seq_name = xml_path.stem
    seq_images = seq_to_images.get(seq_name, [])
    
    if len(seq_images) == 0:
        # fallback: tìm folder ảnh có tên gần giống
        for k, v in seq_to_images.items():
            if k in seq_name or seq_name in k:
                seq_images = v
                break
    
    if len(seq_images) == 0:
        return 0
    
    converted = 0
    
    for frame in root.findall(".//frame"):
        frame_num = int(frame.attrib.get("num", 0))
        img_path = find_image_for_frame(seq_images, frame_num)
        
        if img_path is None or not img_path.exists():
            continue
        
        img = cv2.imread(str(img_path))
        if img is None:
            continue
        
        h_img, w_img = img.shape[:2]
        labels = []
        
        for target in frame.findall(".//target"):
            box = parse_box_from_target(target)
            if box is None:
                continue
            
            left, top, width, height = box
            cls_id = parse_class_from_target(target)
            
            x_center = (left + width / 2) / w_img
            y_center = (top + height / 2) / h_img
            w = width / w_img
            h = height / h_img
            
            # clip về [0, 1]
            x_center = min(max(x_center, 0), 1)
            y_center = min(max(y_center, 0), 1)
            w = min(max(w, 0), 1)
            h = min(max(h, 0), 1)
            
            if w <= 0 or h <= 0:
                continue
            
            labels.append(f"{cls_id} {x_center:.6f} {y_center:.6f} {w:.6f} {h:.6f}")
        
        if len(labels) == 0:
            continue
        
        out_img_name = f"{seq_name}_{img_path.name}"
        out_label_name = Path(out_img_name).with_suffix(".txt").name
        
        out_img_path = IMAGES_DIR / split / out_img_name
        out_label_path = LABELS_DIR / split / out_label_name
        
        shutil.copy2(img_path, out_img_path)
        out_label_path.write_text("\n".join(labels))
        
        converted += 1
    
    return converted
```

```python
# 7) Chia train/val theo sequence rồi convert

random.seed(42)
xml_paths = sorted(xml_paths)

random.shuffle(xml_paths)
val_ratio = 0.2
val_count = int(len(xml_paths) * val_ratio)

val_xml = set(xml_paths[:val_count])
train_xml = set(xml_paths[val_count:])

total_train = 0
total_val = 0

for xml in tqdm(xml_paths):
    if xml in val_xml:
        total_val += convert_xml_to_yolo(xml, split="val")
    else:
        total_train += convert_xml_to_yolo(xml, split="train")

print("Converted train images:", total_train)
print("Converted val images:", total_val)
```

```python
# 10) Train YOLO11n trên GPU Vast.ai

from ultralytics import YOLO

model = YOLO("yolo11n.pt")
yaml_path = YOLO_ROOT / "ua_detrac.yaml"

train_results = model.train(
    data=str(yaml_path),
    epochs=50,
    imgsz=640,
    batch=16,
    device=0,
    workers=4,
    pretrained=True,
    cache=True,
    patience=15,
    project="/CITD/runs",
    name="yolo11n_ua_detrac_4cls"
)
```

```python
# 13) Tạo video từ một sequence ảnh UA-DETRAC

import cv2
from pathlib import Path

# Lấy một folder ảnh bất kỳ trong dataset gốc
seq_folder = None
for folder in DATA_ROOT.rglob("*"):
    if folder.is_dir():
        imgs = sorted(list(folder.glob("*.jpg")) + list(folder.glob("*.png")))
        if len(imgs) > 50:
            seq_folder = folder
            break

print("Sequence folder:", seq_folder)

seq_imgs = sorted(list(seq_folder.glob("*.jpg")) + list(seq_folder.glob("*.png")))

first_img = cv2.imread(str(seq_imgs[0]))
h, w = first_img.shape[:2]

video_path = "/CITD/ua_detrac_sequence_test.mp4"
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
writer = cv2.VideoWriter(video_path, fourcc, 25, (w, h))

for img_path in seq_imgs:
    frame = cv2.imread(str(img_path))
    if frame is None:
        continue
    writer.write(frame)

writer.release()

print("Saved video:", video_path)
```
